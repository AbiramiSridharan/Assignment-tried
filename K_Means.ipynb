```python
from sklearn.cluster import KMeans
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import plotly as py
import plotly.graph_objs as go
import warnings
import seaborn as sns
```

```python
X = np.array([[2,4],[2,6],[5,6],[4,7],[8,3],[6,6],[5,2],[5,7],[6,3],[4,4]])
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
kmeans.labels_
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
kmeans.cluster_centers_
```

```python
X = np.array([[2,10],[2,5],[8,4],[5,8],[7,5],[6,4],[1,2],[4,9]])
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
kmeans.labels_
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
kmeans.cluster_centers_
```

```python
x1=data[['Age','Spending Score (1-100)']].iloc[:,:].values
inertia=[]
for n in range(1,11):
    algorithm=(KMeans(n_clusters=n,init='k-means++',n_init=10,max_iter=300,random_state=111,algorithm='elkan'))
    algorithm.fit(x1)
    inertia.append(algorithm.inertia_)
```

```python
plt.figure(1,figsize=(15,6))
plt.plot(np.arange(1,11),inertia,'o')
plt.plot(np.arange(1,11),inertia,'-',alpha=0.5)
plt.xlabel('Number of clusters'),plt.ylabel('Inertia')
plt.show()
```

```python
algorithm=KMeans(n_clusters=4,init='k-means++',n_init=10,max_iter=300,tol=0.0001,random_state=111,algorithm='elkan')
algorithm.fit(x1)
labels1=algorithm.labels_
centroids1=algorithm.cluster_centers_
```

```python
plt.scatter(data['Age'],data['Spending Score (1-100)'], c=algorithm.labels_)
algorithm.cluster_centers_
```
